## 1. Problem Overview.

A Terry stop is a brief detention or stop of a person by police on reasonable suspicion of involvement in criminal activity but short of probable cause to arrest. This practice originates from the U.S. Supreme Court case Terry v. Ohio (1968).

Datasets related to Terry stops typically include information on stop-and-frisk incidents conducted by police. These datasets often contain details such as:

* Date and time of the stop
* Location of the stop
* Demographics of the individuals stopped (e.g., age, gender, race)
* Reason for the stop
* Outcome of the stop (e.g., no action, citation, arrest)
* Whether a frisk or search was conducted
* Contraband or weapons found

Evaluating the effectiveness of Terry stops and understanding racial and financial disparities involves analyzing various aspects of the data to assess how these stops impact different communities and whether they achieve their intended goals. By comprehensively evaluating these factors, we can gain a deeper understanding of the effectiveness of Terry stops and the disparities they may introduce or exacerbate. This analysis is crucial for informing policy decisions and promoting fair and just law enforcement practices.

In this analysis we'll explore the effects of racial and financial disparities in Seattle communities on Terry Stops. 

*One note about these data is important.  The Seattle Police Department does not use "stop and frisk" as a deployment tactic, nor does the Department view the number of stops as indicative of a goal by which to measure proactive policing. Without question, investigative stops, when supported by reasonable suspicion, are a useful tool to address potential criminal activity encountered by police officers, but increasing or even maintaining the level of stops year to year is not a goal for the department.  These data should not, accordingly, be viewed as a proxy for measuring officer proactivity.

We'll be creating a dataset matching the following geographic identificators: `beat`, `TRACTCE10`, `GEOID10`.

```python
# define quiet pip install 
import subprocess
import sys

# Function to run pip install without showing "Requirement already satisfied" messages
def quiet_pip_install(package):
    subprocess.call([sys.executable, "-m", "pip", "install", package], \
                    stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL)

quiet_pip_install('geopandas')
# importing libraries
import warnings
import geopandas as gpd
import pandas as pd
import numpy as np

# file system modules 
import os

#omit warnings
warnings.filterwarnings('ignore')
 

# define directories
beats_dir_data = './data/SEA_Beats'
census_dir_data = './data/KC_Tracts' 

dir_data = './data'
beats_tracts_dataset = 'SEA_Beats_Tracts_Neighborhoods.csv'
dir_files = [f for f in os.listdir(dir_data) if not f.startswith('.')]
    
# print the contents
print('All files: \n', dir_files)

# creating filesystem path to data dir
path_dir_data = os.path.join(dir_data)

# if SEA_Beats_Tracts_Neighborhoods is missing, attempt to re-create and save it
if beats_tracts_dataset not in dir_files:
    print(f"Missing file: {beats_tracts_dataset}")
    # load the police beats shapefile
    police_beats = gpd.read_file(os.path.join(beats_dir_data, 'Beats_WM.shp'))
    # load the census tracts shapefile
    census_tracts = gpd.read_file(os.path.join(census_dir_data,'kc_tract_10.shp'))

    # ensure both datasets use the same coordinate reference system (CRS)
    police_beats = police_beats.to_crs(census_tracts.crs)

    # perform the spatial join
    joined = gpd.sjoin(police_beats, census_tracts, how='left', op='intersects')

    # extract relevant columns
    cols_of_interest = ['beat', 'TRACTCE10', 'GEOID10']
    tract_beat_data = joined[cols_of_interest]
    tract_beat_data.dropna(inplace=True)
    # save to a CSV file
    output_path = os.path.join(path_dir_data, beats_tracts_dataset)
    tract_beat_data.to_csv(output_path, index=False)
else:
    print(f"Found file: {beats_tracts_dataset}")
    data_path = os.path.join(path_dir_data, beats_tracts_dataset)
    tract_beat_data = pd.read_csv(data_path)    

tract_beat_data.head(20)
```

### Interpreting the index values.

Before we continue, it's very important to understand what our columns of interest contain andf how to interpret the numbers. Below we'll give a brief overoverview of what the RSE index is and how it is composed.
#### What it is.
The RSE index was created in 2017 by  City Demographer in the Office of Planning & Community Development (OPCD) with involvement of multiple departments.The RSE Index is a tool to aid in the identification of geographic priorities for City programs, planning efforts, and investments

#### What composes the index.
The RSE Index is a composite index made up of three equally weighted sub-indices, as shown below with their component metrics:

*  Race, ELL, and Origins Index 
    − Persons of color 
    − English language learners (ELL)* 
    − Foreign born* (*Weighted by 0.5 in subindex) 
    
* Socioeconomic Disadvantage Index 
    − Income below 200 percent of poverty 
    − Educational attainment less than a bachelor’s degree 
    
* Health Disadvantage Index 
    − Adults w/no leisure-time physical activity 
    − Adults with diagnosed diabetes 
    − Adults with obesity 
    − Adults who report that their mental health is not good 
    − Adults with asthma 
    − Low life expectancy at birth 
    − Adults with one or more disabilities
        
#### How to interpret each sub-index.
* Race, ELL, and Origins Index: A higher score indicates areas with more racial and ethnic diversity, higher numbers of English language learners, and a greater proportion of foreign-born residents.

* Socioeconomic Disadvantage Index: A higher score indicates areas with higher levels of socioeconomic disadvantage, such as lower income, lower education levels, and higher unemployment rates.
* Health Disadvantage Index: A higher score indicates areas with greater health disadvantages, including poorer health outcomes and lower access to healthcare services.

#### Example: calculating Race, ELL, and Origins Index for Tract 110.02.

- **People-of-color metric (Weight: 1.0):**
  - Percentile for people of color: 1.00 (100th percentile)

- **ELL (English Language Learners) metric (Weight: 0.5):**
  - Percentile for ELL: 0.99 (99th percentile)

- **Foreign-born metric (Weight: 0.5):**
  - Percentile for foreign-born: 0.98 (98th percentile)

##### Formula:  Sub-index score = (1 * Percentile for people of color) + (0.5 * Percentile for ELL) + (0.5 * Percentile for foreign-born)

- **Calculation for Tract 110.02:** 
    Sub-index score = (1 * 1.00) + (0.5 * 0.99) + (0.5 * 0.98) = 0.994

#### Adding Officer's Age Group Column

```python
# Convert 'Reported Date' to datetime format
terry_subset['Reported Date'] = pd.to_datetime(terry_subset['Reported Date'])

# Extract year from 'Reported Date' as current year
current_year = terry_subset['Reported Date'].dt.year

# Calculate the officer's age based on the current year
terry_subset['Officer Age'] = current_year - terry_subset['Officer YOB']

# Define the age_group function
def age_group(age):
    if age < 16:
        return 'Below 16'
    elif age <= 26:
        return '16-25'
    elif age <= 35:
        return '26-35'
    elif age <= 45:
        return '36-45'
    elif age <= 55:
        return '46-55'
    else:
        return '55+'

# Apply the function to the 'Officer Age' column
terry_subset['Officer Age Group'] = terry_subset['Officer Age'].apply(age_group)

# Drop the 'Officer Age' column if you don't need it
terry_subset.drop(columns=['Officer Age'], inplace=True)

# Display the updated DataFrame
terry_subset.head()
```

#### Adding Categorical Weekday/Weekend column

```python
# Convert 'Reported Date' to datetime format
terry_subset['Reported Date'] = pd.to_datetime(terry_subset['Reported Date'])

# Add a column to indicate weekday or weekend
terry_subset['Reported Day'] = terry_subset['Reported Date'].dt.day_name()
terry_subset['Weekday'] = terry_subset['Reported Date'].dt.dayofweek.map({0:1, \
                                                                          1:1, 2:1, 3:1, \
                                                                          4:1, 5:0, 6:0})

# Display the updated DataFrame
terry_subset.head()
```

#### Adding Categorical Time of Day Column

```python
# Ensure 'Reported Time' is a string, then remove seconds and fractional seconds
terry_subset['Reported Time'] = terry_subset['Reported Time'].astype(str).str.slice(0, 5)
terry_subset['Reported Time'] = pd.to_datetime(terry_subset['Reported Time'], format='%H:%M').dt.time

# Function to categorize time
def categorize_time(time):
    if time >= pd.to_datetime('00:00', format='%H:%M').time() and time < pd.to_datetime('04:00', format='%H:%M').time():
        return 'late night'
    elif time >= pd.to_datetime('04:00', format='%H:%M').time() and time < pd.to_datetime('08:00', format='%H:%M').time():
        return 'early morning'
    elif time >= pd.to_datetime('08:00', format='%H:%M').time() and time < pd.to_datetime('12:00', format='%H:%M').time():
        return 'morning'
    elif time >= pd.to_datetime('12:00', format='%H:%M').time() and time < pd.to_datetime('16:00', format='%H:%M').time():
        return 'afternoon'
    elif time >= pd.to_datetime('16:00', format='%H:%M').time() and time < pd.to_datetime('20:00', format='%H:%M').time():
        return 'evening'
    else:
        return 'early night'

# Apply the function to the 'Reported Time' column
terry_subset['Time of Day'] = terry_subset['Reported Time'].apply(categorize_time)

# Display the updated DataFrame
terry_subset.head()
```

## 7. Improving the basic classifier.

We'll switch to `LogisticRegression` from `sklearn.linear_model` so we can use a different model solver and have more parameter options. We'll also switch training-testing sample sizes from 70%-30% to 75%-25%.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.preprocessing import StandardScaler

X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.25, random_state=42)

# treate a logistic regression model with L2 regularization (Lasso)
lasso_model = LogisticRegression(fit_intercept=True, penalty='l1', solver='saga', C=100.0, max_iter=1000)
# train the model
lasso_model.fit(X_train, y_train)
```

```python
# Coefficients
print("Coefficients:",  lasso_model.coef_)
# Intercept
print("Intercept:", lasso_model.intercept_)
```

### Model performance, confusion matrix, ROC.

Here we'll look into how the classifier performs with the default parameters.

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

y_hat_train = lasso_model.predict(X_train)
y_hat_test = lasso_model.predict(X_test)

#look at the residuals and calculate the MSE for training and test sets:

train_mse = mean_squared_error(y_train, y_hat_train)
test_mse = mean_squared_error(y_test, y_hat_test)
print('Train Mean Squared Error:', train_mse)
print('Test Mean Squared Error:', test_mse)
# Print classification report:
print(classification_report(y_test, y_hat_test))

# Confusion Matrix for Logistic Regression:
cm_LR = confusion_matrix(y_hat_test, y_test)
disp = ConfusionMatrixDisplay(confusion_matrix=cm_LR, display_labels=['Arrested', 'Not Arrested'])
disp.plot(cmap=plt.cm.Blues)
plt.title("Confusion Matrix Lasso")
plt.show()
```

```python
from sklearn.metrics import roc_curve, auc
y_train_score = lasso_model.decision_function(X_train)
y_test_score = lasso_model.decision_function(X_test)

train_fpr, train_tpr, train_thresholds = roc_curve(y_train, y_train_score)
test_fpr, test_tpr, test_thresholds = roc_curve(y_test, y_test_score)

print('Train AUC: {}'.format(auc(train_fpr, train_tpr)))
print('Test AUC: {}'.format(auc(test_fpr, test_tpr)))

plt.figure(figsize=(10, 8))
lw = 2

plt.plot(train_fpr, train_tpr, color='blue',
         lw=lw, label='Train ROC curve')
plt.plot(test_fpr, test_tpr, color='darkorange',
         lw=lw, label='Test ROC curve')

plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.yticks([i/20.0 for i in range(21)])
plt.xticks([i/20.0 for i in range(21)])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic (ROC) Curve')
plt.legend(loc='lower right')
plt.show()
```

## Exploring the effects of paramters on model performance.
In this section we'll explore the parameters space to see if we can improve the performance. We'll check the constant C and train-test split size.

### Iterate over possible train-test sample sizes.

We'll use different train-test split sizes, for each size run classifier, and store presicion, f1, recall metrics for each split. We'll use same random state each time.

```python
from sklearn.metrics import roc_curve, auc, accuracy_score, f1_score, recall_score, precision_score

training_precision = []
testing_precision = []
training_recall = []
testing_recall = []
training_accuracy = []
testing_accuracy = []
training_f1 = []
testing_f1 = []

for i in range(10, 95):
    X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, \
                                                        test_size=i/100.0, random_state=42)
    logreg = LogisticRegression(fit_intercept=True, C=100, solver='saga', penalty='l1')
    model_log = logreg.fit(X_train, y_train)
    y_hat_test = logreg.predict(X_test)
    y_hat_train = logreg.predict(X_train)

    training_precision.append(precision_score(y_train, y_hat_train))
    testing_precision.append(precision_score(y_test, y_hat_test))
    training_recall.append(recall_score(y_train, y_hat_train))
    testing_recall.append(recall_score(y_test, y_hat_test))
    training_accuracy.append(accuracy_score(y_train, y_hat_train))
    testing_accuracy.append(accuracy_score(y_test, y_hat_test))
    training_f1.append(f1_score(y_train, y_hat_train))
    testing_f1.append(f1_score(y_test, y_hat_test))
```

### Plotting the performace metrics. 

```python
# Train and test precision
plt.scatter(list(range(10, 95)), training_precision, label='training_precision')
plt.scatter(list(range(10, 95)), testing_precision, label='testing_precision')
plt.legend()
plt.show()
```

```python
# Train and test recall
plt.scatter(list(range(10, 95)), training_recall, label='training_recall')
plt.scatter(list(range(10, 95)), testing_recall, label='testing_recall')
plt.legend()
plt.show()
```

```python
# Train and test accuracy
plt.scatter(list(range(10, 95)), training_accuracy, label='training_accuracy')
plt.scatter(list(range(10, 95)), testing_accuracy, label='testing_accuracy')
plt.legend()
plt.show()
```

```python
# Train and test F1 score
plt.scatter(list(range(10, 95)), training_f1, label='training_f1')
plt.scatter(list(range(10, 95)), testing_f1, label='testing_f1')
plt.legend()
plt.show()
```

### Conclusion on train-test parameter size.
We tested train-test splits ranging from 0.1 to 0.95. Looking at clasifier performance for each of them, there is a swet spot between 0.6-0.8 but overall the gains are marginal.  

### Exploring the Constant value.
Here we'll be looking into classifier performance for various values of parameter `C`.
We'll plot the ROC curve for each and compare AUC values.

```python
fig, axes = plt.subplots(4,2, figsize=(15, 15))
for n in range(8):
    i = n%4
    j = n//4
    ax = axes[i, j]
    # Fit a model
    logreg = LogisticRegression(fit_intercept=True, C=1.5**(n), solver='saga', penalty='l1')
    logreg.fit(X_train, y_train)
    
    y_test_score = logreg.decision_function(X_test)
    y_train_score = logreg.decision_function(X_train)

    test_fpr, test_tpr, test_thresholds = roc_curve(y_test, y_test_score)
    train_fpr, train_tpr, train_thresholds = roc_curve(y_train, y_train_score)
    
    test_auc = auc(test_fpr, test_tpr)
    train_auc = auc(train_fpr, train_tpr)
    auc_diff = round(train_auc - test_auc, 4)

    print('Test AUC with C=1.5^{}: {}'.format(n*2, auc(test_fpr, test_tpr)))
    print('Train AUCwith C=1.5^{}: {}'.format(n*2, auc(train_fpr, train_tpr)))
    # Add the plot
    ax.plot(test_fpr, test_tpr, color='darkorange',
         lw=lw, label='Test ROC curve')
    ax.plot(train_fpr, train_tpr, color='blue',
             lw=lw, label='Train ROC curve')
    
    ax.set_title('Regularization Parameter set to: 1.5^{}\nDifference in Test/Train AUC: {}'.format(n, auc_diff))
```
