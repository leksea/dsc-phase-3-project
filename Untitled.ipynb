### Terry Stop definition.
A Terry stop is a brief detention or stop of a person by police on reasonable suspicion of involvement in criminal activity but short of probable cause to arrest. This practice originates from the U.S. Supreme Court case Terry v. Ohio (1968).

Datasets related to Terry stops typically include information on stop-and-frisk incidents conducted by police. These datasets often contain details such as:

* Date and time of the stop
* Location of the stop
* Demographics of the individuals stopped (e.g., age, gender, race)
* Reason for the stop
* Outcome of the stop (e.g., no action, citation, arrest)
* Whether a frisk or search was conducted
* Contraband or weapons found

Evaluating the effectiveness of Terry stops and understanding racial and financial disparities involves analyzing various aspects of the data to assess how these stops impact different communities and whether they achieve their intended goals. By comprehensively evaluating these factors, we can gain a deeper understanding of the effectiveness of Terry stops and the disparities they may introduce or exacerbate. This analysis is crucial for informing policy decisions and promoting fair and just law enforcement practices.

In this analysis we'll explore the effects of racial and financial disparities in Seattle communities on Terry Stops. 

*One note about these data is important.  The Seattle Police Department does not use "stop and frisk" as a deployment tactic, nor does the Department view the number of stops as indicative of a goal by which to measure proactive policing. Without question, investigative stops, when supported by reasonable suspicion, are a useful tool to address potential criminal activity encountered by police officers, but increasing or even maintaining the level of stops year to year is not a goal for the department.  These data should not, accordingly, be viewed as a proxy for measuring officer proactivity.

### Interpreting the index values.

Before we continue, it's very important to understand what our columns of interest contain andf how to interpret the numbers. Below we'll give a brief overoverview of what the RSE index is and how it is composed.
#### What it is.
The RSE index was created in 2017 by  City Demographer in the Office of Planning & Community Development (OPCD) with involvement of multiple departments.The RSE Index is a tool to aid in the identification of geographic priorities for City programs, planning efforts, and investments

#### What composes the index.
The RSE Index is a composite index made up of three equally weighted sub-indices, as shown below with their component metrics:

*  Race, ELL, and Origins Index 
    − Persons of color 
    − English language learners (ELL)* 
    − Foreign born* (*Weighted by 0.5 in subindex) 
    
* Socioeconomic Disadvantage Index 
    − Income below 200 percent of poverty 
    − Educational attainment less than a bachelor’s degree 
    
* Health Disadvantage Index 
    − Adults w/no leisure-time physical activity 
    − Adults with diagnosed diabetes 
    − Adults with obesity 
    − Adults who report that their mental health is not good 
    − Adults with asthma 
    − Low life expectancy at birth 
    − Adults with one or more disabilities
        
#### How to interpret each sub-index.
* Race, ELL, and Origins Index: A higher score indicates areas with more racial and ethnic diversity, higher numbers of English language learners, and a greater proportion of foreign-born residents.

* Socioeconomic Disadvantage Index: A higher score indicates areas with higher levels of socioeconomic disadvantage, such as lower income, lower education levels, and higher unemployment rates.
* Health Disadvantage Index: A higher score indicates areas with greater health disadvantages, including poorer health outcomes and lower access to healthcare services.

#### Example: calculating Race, ELL, and Origins Index for Tract 110.02.

- **People-of-color metric (Weight: 1.0):**
  - Percentile for people of color: 1.00 (100th percentile)

- **ELL (English Language Learners) metric (Weight: 0.5):**
  - Percentile for ELL: 0.99 (99th percentile)

- **Foreign-born metric (Weight: 0.5):**
  - Percentile for foreign-born: 0.98 (98th percentile)

##### Formula:  Sub-index score = (1 * Percentile for people of color) + (0.5 * Percentile for ELL) + (0.5 * Percentile for foreign-born)

- **Calculation for Tract 110.02:** 
    Sub-index score = (1 * 1.00) + (0.5 * 0.99) + (0.5 * 0.98) = 0.994

#### Adding Officer's Age Group Column.

```python
# convert 'Reported Date' to datetime format
terry_subset['Reported Date'] = pd.to_datetime(terry_subset['Reported Date'])

# extract year from 'Reported Date' as current year
current_year = terry_subset['Reported Date'].dt.year

# calculate the officer's age based on the current year
terry_subset['Officer Age'] = current_year - terry_subset['Officer YOB']

# define the age_group function
def age_group(age):
    if age < 16:
        return 'Below 16'
    elif age <= 26:
        return '16-25'
    elif age <= 35:
        return '26-35'
    elif age <= 45:
        return '36-45'
    elif age <= 55:
        return '46-55'
    else:
        return '55+'

# apply the function to the 'Officer Age' column
terry_subset['Officer Age Group'] = terry_subset['Officer Age'].apply(age_group)

# drop the 'Officer Age' column if you don't need it
terry_subset.drop(columns=['Officer Age'], inplace=True)

# display the updated DataFrame
terry_subset['Officer Age Group'].value_counts()
```

#### Adding Categorical Weekday/Weekend column.

```python
# convert 'Reported Date' to datetime format
terry_subset['Reported Date'] = pd.to_datetime(terry_subset['Reported Date'])

# add a column to indicate weekday or weekend
terry_subset['Reported Day'] = terry_subset['Reported Date'].dt.day_name()
terry_subset['Weekday'] = terry_subset['Reported Date'].dt.dayofweek.map({0:1, \
                                                                          1:1, 2:1, 3:1, \
                                                                          4:1, 5:0, 6:0})

# display the updated DataFrame
terry_subset['Weekday'].value_counts()
```

#### Adding Categorical Time of Day Column

```python
# Ensure 'Reported Time' is a string, then remove seconds and fractional seconds
terry_subset['Reported Time'] = terry_subset['Reported Time'].astype(str).str.slice(0, 5)
terry_subset['Reported Time'] = pd.to_datetime(terry_subset['Reported Time'], format='%H:%M').dt.time

# Function to categorize time
def categorize_time(time):
    if time >= pd.to_datetime('00:00', format='%H:%M').time() and time < pd.to_datetime('04:00', format='%H:%M').time():
        return 'late night'
    elif time >= pd.to_datetime('04:00', format='%H:%M').time() and time < pd.to_datetime('08:00', format='%H:%M').time():
        return 'early morning'
    elif time >= pd.to_datetime('08:00', format='%H:%M').time() and time < pd.to_datetime('12:00', format='%H:%M').time():
        return 'morning'
    elif time >= pd.to_datetime('12:00', format='%H:%M').time() and time < pd.to_datetime('16:00', format='%H:%M').time():
        return 'afternoon'
    elif time >= pd.to_datetime('16:00', format='%H:%M').time() and time < pd.to_datetime('20:00', format='%H:%M').time():
        return 'evening'
    else:
        return 'early night'

# Apply the function to the 'Reported Time' column
terry_subset['Time of Day'] = terry_subset['Reported Time'].apply(categorize_time)

# Display the updated DataFrame
terry_subset['Time of Day'].value_counts()
```

## 7. Improving the basic classifier.

We'll utilize the `Pipeline` library to create an automated data processing ppeline.
We'll utilize the `GridSearchCV` for parameter optimization and cross-validation.
We won't be using train-test split as the the dataset will be split into training and validation subsets multiple times to evaluate the performance of different parameter combinations.

### Selecting the scoring:

Our data is imbalanced and using accuracy as an optimization metric would be misleading.

If we use presicion, we'll have the high cost of false positives. For example, if labeling someone as an arrest when they are not is particularly harmful or resource-intensive.

If we use recall, we'll have a high cost of false negatives. For example, if missing an actual arrest is particularly harmful (e.g., letting a potential offender go free).

Since both are important, we'll use the f1 metric to balance precision and recall.

```python
# importing Pipeline, GridSearchCV
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from imblearn.pipeline import Pipeline as ImbPipeline
from sklearn.metrics import make_scorer, f1_score

# re-defining the data
X = data.drop('Arrest Flag', axis=1)
y = data['Arrest Flag']

# define the preprocessor with OneHotEncoder for categorical columns and StandardScaler for continuous columns
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(sparse=False, drop='first'), categorical_columns),
        ('num', StandardScaler(), continuous_columns)
    ])

# define the pipeline with SMOTE and Logistic Regression
pipeline = ImbPipeline(steps=[
    ('preprocessor', preprocessor),
    ('smote', SMOTE(random_state=42)),
    ('classifier', LogisticRegression(fit_intercept=True, penalty='l1', solver='saga', max_iter=1000))
])

# define the parameter grid
param_grid = {
    'classifier__C': [0.1, 1, 10, 100],  # Regularization strength
    'classifier__max_iter': [500, 1000, 1500],  # Number of iterations
    'classifier__solver': ['liblinear', 'saga']  # Solvers for Logistic Regression
}

# define the f1_scorer
f1_scorer = make_scorer(f1_score)

# initialize GridSearchCV
grid_search_logreg = GridSearchCV(pipeline, param_grid, cv=5, scoring=f1_scorer)

# split into train-test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42) 

# fit the grid search to the data
grid_search_logreg.fit(X_train, y_train)

# print the best parameters and the best score
print("Best parameters found: ", grid_search_logreg.best_params_)
print("Best cross-validation score: ", grid_search_logreg.best_score_)
```

### Optimized classifier performance.

We'll run classification report and plot confusion matrix.

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# get the best estimator
best_model = grid_search_logreg.best_estimator_

# make predictions on the test set we defined earlier
y_pred = best_model.predict(X_test)

# compute the confusion matrix
cm = confusion_matrix(y_test, y_pred)

# visualize the confusion matrix
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=best_model.classes_)
disp.plot()

# print the classification report
print("Classification Report:")
print(classification_report(y_test, y_pred))
```
